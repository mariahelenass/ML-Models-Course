```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
transacoes_lidas = []

with open("transacoes_produtos.txt", "r") as arquivo:
    for linha in arquivo:
        produtos_transacao = linha.strip().split(", ")
        transacoes_lidas.append(produtos_transacao)
```

```python
transacoes = TransactionEncoder()
transacoes_array = transacoes.fit(transacoes_lidas).transform(transacoes_lidas)
```

```python
df = pd.DataFrame(transacoes_array, columns=transacoes.columns_)
```

```python
df
```

```python
# itens frequentes
frequent_items = apriori(df, min_support=0.5, use_colnames=True)
frequent_items
```

```python
# min_threshold - minimo de confiança
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.5)
print(rules)
```
