```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/Customer_Behaviour.csv')
```

```python
df.head()
```

```python
df = df.drop(columns='User ID', axis=1)
```

```python
df = pd.get_dummies(df, columns=['Gender'])
```

```python
df.head()
```

```python
X = df.drop(columns='Purchased', axis=1)
y = df['Purchased']
```

```python
X.head()
```

```python
sc = StandardScaler()
dados_norm = sc.fit_transform(X.iloc[:,0:2])
```

```python
dados_norm_df = pd.DataFrame(dados_norm, columns=['Age_norm', 'EstimatedSalary_norm'])

X_novo = pd.concat([X, dados_norm_df], axis=1)
```

```python
X_novo.head()
```

```python
colunas = ['Age', 'EstimatedSalary', 'Gender_Female']

X_novo.drop(columns=colunas, axis=1)
```
